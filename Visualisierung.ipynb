```python
import pandas as pd
import numpy as np
from datetime import datetime 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_csv = pd.read_csv("Daten_Aufgabe_1 - Shopauswertung.csv",  delimiter=";")

df_csv.head(3)
```

```python
sns.barplot(x=df_csv.index, y=df_csv['Umsatz'], data=df_csv)

# Show the plot
plt.show()
```

```python
df_csv['Datum'] = pd.to_datetime(df_csv.Datum, format='%d.%m.%Y')
```

```python
fig, ax1 = plt.subplots(figsize=(15,10))
sns.lineplot(x=df_csv.index, y=df_csv['Umsatz'], data=df_csv)
# Show the plot
plt.show()
```

```python
cleaned_inland_excel = pd.read_excel("cleaned_data_inland.xlsx")

cleaned_inland_excel.head(3)
```

```python
fig, ax1 = plt.subplots(figsize=(15,10))
sns.lineplot(x=cleaned_inland_excel.index, y=(cleaned_inland_excel['Produkt_A_Inland']))
sns.lineplot(x=cleaned_inland_excel.index, y=(cleaned_inland_excel['Produkt_B_Inland']))
sns.lineplot(x=cleaned_inland_excel.index, y=(cleaned_inland_excel['Produkt_C_Inland']))
# Show the plot
plt.show()
```

```python
cleaned_inland_excel_sum = pd.read_excel("cleaned_data_inland_sum.xlsx") # cleaned_data_inland_sum.xlsx

cleaned_inland_excel_sum.head(3)
```

```python
fig, ax1 = plt.subplots(figsize=(15,10))
sns.lineplot(x=cleaned_inland_excel_sum.index, y=(cleaned_inland_excel_sum['Produkt_A_Inland_sum']))
sns.lineplot(x=cleaned_inland_excel_sum.index, y=(cleaned_inland_excel_sum['Produkt_B_Inland_sum']))
sns.lineplot(x=cleaned_inland_excel_sum.index, y=(cleaned_inland_excel_sum['Produkt_C_Inland_sum']))
# Show the plot
plt.show()
```

```python
cleaned_inland_excel[['Produkt_A_Inland', 'Produkt_B_Inland', 'Produkt_C_Inland']].boxplot()
```

```python
cleaned_inland_excel[['Produkt_C_Inland', 'Produkt_B_Inland', 'Produkt_A_Inland']].plot(kind='bar', stacked=True)
```

```python
cleaned_inland_excel.plot(cleaned_inland_excel.Produkt_A_Inland, cleaned_inland_excel.Uptrend, label="uptrend")
cleaned_inland_excel.plot(cleaned_inland_excel.Produkt_A_Inland, cleaned_inland_excel.Downtrend, label="downtrend")

plt.legend()
plt.grid()
plt.show()
```
