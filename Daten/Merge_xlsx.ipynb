```python
import numpy as np
import pandas as pd
import re
```

```python
df_d = pd.read_excel(r"C:\Users\User\Documents\GitHub\Praxisprojekt_Modul_2.2_Breakout2\Daten\Daten_Aufgabe_1 - Shop Analyse.xlsx",sheet_name=["P1","P2", "P3"])
df_d
```

```python
a= df_d["P1"].values
df_A = pd.DataFrame(a, columns=["Monat", "Produkt_A_Inland"])
df_A
```

```python
b= df_d["P2"].values
df_B = pd.DataFrame(b, columns=["Monat", "Produkt_B_Inland"])
df_B
```

```python
c= df_d["P3"].values
df_C = pd.DataFrame(c, columns=["Monat", "Produkt_C_Inland"])
df_C
```

```python
df_inland= df_A.merge(df_B)
```

```python
df_inland = df_inland.merge(df_C)
```

```python
df_inland
```

```python
df_inland.set_index("Monat", inplace=True)
```

```python
df_inland
```

```python
df_inland["Umsatz_A_Inland"]=df_inland["Produkt_A_Inland"].astype("float").round(2)
df_inland["Umsatz_B_Inland"]=df_inland["Produkt_B_Inland"].astype("float").round(2)
df_inland["Umsatz_C_Inland"]=df_inland["Produkt_C_Inland"].astype("float").round(2)
```

```python
df_inland = df_inland.drop(["Produkt_A_Inland", "Produkt_B_Inland", "Produkt_C_Inland"], axis=1)
 
```

```python
df_inland["Umsatz_gesamt_Inland"] = df_inland["Umsatz_A_Inland"] + df_inland["Umsatz_B_Inland"] + df_inland["Umsatz_C_Inland"]
```
