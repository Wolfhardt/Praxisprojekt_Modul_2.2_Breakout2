```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\User\Documents\GitHub\Praxisprojekt_Modul_2.2_Breakout2\Daten\Umsatz_Ausland_Tage.csv",usecols=["Datum", "Umsatz_gesamt_Ausland", "Umsatz_A_Ausland", "Umsatz_B_Ausland", "Umsatz_C_Ausland"])
df
```

```python
df["Umsatz_B_Ausland"] = df["Umsatz_B_Ausland"].diff().fillna(df["Umsatz_B_Ausland"][0])
df["Umsatz_gesamt_Ausland"] = df["Umsatz_gesamt_Ausland"] - df["Umsatz_B_Ausland"] - df["Umsatz_A_Ausland"]
df["Umsatz_gesamt_Ausland"] = df["Umsatz_gesamt_Ausland"].diff().fillna(0)
df["Umsatz_gesamt_Ausland"][0] = df["Umsatz_gesamt_Ausland"].median()
df = df.rename(columns={"Umsatz_gesamt_Ausland":"Umsatz_C_Ausland", "Umsatz_C_Ausland":"Umsatz_gesamt_Ausland"})
#df["Umsatz_gesamt_Ausland"] = df["Umsatz_A_Ausland"] + df["Umsatz_B_Ausland"] + df["Umsatz_C_Ausland"]
df
```

```python
df.dtypes
```

```python
df["Umsatz_A_Ausland"]=df["Umsatz_A_Ausland"].round(2)
df["Umsatz_B_Ausland"]=df["Umsatz_B_Ausland"].round(2)
df["Umsatz_C_Ausland"]=df["Umsatz_C_Ausland"].round(2)
#df["Umsatz_gesamt_Ausland"]=["Umsatz_gesamt_Ausland"].round(2)
df
```

```python
df["Umsatz_gesamt_Ausland"] = df["Umsatz_A_Ausland"] + df["Umsatz_B_Ausland"] + df["Umsatz_C_Ausland"]
df
```

```python
df = pd.read_csv(r"C:\Users\User\Documents\GitHub\Praxisprojekt_Modul_2.2_Breakout2\Daten\Umsatz_Ausland_Tage.csv",usecols=["Datum", "Umsatz_gesamt_Ausland", "Umsatz_A_Ausland", "Umsatz_B_Ausland", "Umsatz_C_Ausland"])
df.head(50)
```

```python
df["Datum"] = pd.to_datetime(df["Datum"])
df_monat= df.groupby(pd.Grouper(key="Datum", freq="M")).sum()
df_monat
```

```python
df_monat = df_monat.reset_index()
```

```python
df_monat["Datum"] = df_monat["Datum"].dt.strftime("%b")
df_monat
```

```python
df_monat.to_csv(r"C:\Users\User\Documents\GitHub\Praxisprojekt_Modul_2.2_Breakout2\Daten\Umsatz_Ausland_Monat.csv")
```
